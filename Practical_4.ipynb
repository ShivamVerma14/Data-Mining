## Q4. Run Apriori algorithm to find frequent itemsets and association rules.

```python
!pip install mlxtend
```

```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = [
    ['A', 'B', 'C', 'D', 'F', 'H'],
    ['B', 'E', 'F', 'H'],
    ['A', 'C', 'E'],
    ['B', 'C', 'D', 'F', 'H'],
    ['A', 'B', 'C', 'D', 'E'],
    ['C','D','F','H'],
    ['A','C','D','H'],
    ['E','H']
]
```

```python
encoder = TransactionEncoder()
transactions = encoder.fit_transform(dataset)

data = pd.DataFrame(transactions, columns=encoder.columns_)
data
```

### 4.1	Use minimum support as 50% and minimum confidence as 75%.

```python
frequent_itemsets = apriori(data, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
association_rules(frequent_itemsets, metric='confidence', min_threshold=0.75)
```

### 4.2 Use minimum support as 60% and minimum confidence as 60 %.

```python
frequent_itemsets = apriori(data, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
```

## Apply apriori algorithm and find association rules on Grocery Purchase Dataset.

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

**Importing the dataset.**

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
path = '/content/drive/MyDrive/College Work/Data Mining/Grocery Products Purchase.csv'

data = pd.read_csv(path)
data.head()
```

**Getting sparse matrix of transactions.**

```python
transactions = pd.get_dummies(data.unstack().dropna()).groupby(level=1).sum()
transactions.head()
```

**Minimum Support = 0.1**

```python
frequent_itemsets = apriori(transactions, min_support=0.1, use_colnames=True)
frequent_itemsets
```
